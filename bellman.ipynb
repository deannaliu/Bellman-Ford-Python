```python
import array
import numpy as np

keys = list(); #FROM
values = list(); #TO
distances = list(); #HOW FAR
        
print("Enter name of file to read from:")
f = input()
file = open(f)

#print (matrix)
for i in file:
    key, value, distance = i.rsplit(',')
    keys.append(key)
    values.append(value)
    distances.append(distance.split('\n')[0])

def remove(dupes): 
    finalList = [] 
    for num in dupes: 
        if num not in finalList: 
            finalList.append(num) 
    return finalList 

nodes = ['0', 'x']
noDupes = remove(keys+values)
noDupes2 = remove(keys+values)
noDupes.sort()
noDupes2.sort()
noZero = ['x']
for i in range(len(noDupes)):
    if(not noDupes[i] == 'x' and not noDupes[i] == 'y'):
        nodes.extend(noDupes[i]);
        noZero.extend(noDupes2[i]);
        
nodes.extend('y')
noZero.extend('y')

w, h = len(nodes), len(nodes);
matrix = [[0 for x in range(w)] for y in range(h)] #hold each path

matrix[0] = nodes;

for i in range(len(matrix)):
    matrix[i][0] = nodes[i] #each row starts with the node which is the nodes TO

for i in range(len(matrix)):
    if(i != 0):
        for j in range(len(matrix[i])):
            if(j != 0):
                matrix[i][j] = noZero[j-1]
    
matrix = np.array(matrix)
np.fill_diagonal(matrix, 0, False)

#print(matrix)

matrix=matrix.tolist();

for k in range(len(keys)):
    for i in range(len(matrix)): #go thru each ROW of the matrix
        if(i != 0): # Skip the first 0
            for j in range(len(matrix[i])): #go thru each COLUMN of the matrix
                if(j != 0): #skip the first value in the column
                    if(matrix[i][0] == keys[k] and matrix[i][j] == values[k]):
                        matrix[i][j] = distances[k]
                
for i in range(len(matrix)):
    if(i != 0):
        for j in range(len(matrix[i])): #go thru each COLUMN of the matrix
                if(j != 0): #skip the first value in the column
                    if(matrix[i][j].isnumeric()):
                        matrix[i][j] = int(matrix[i][j])
                    else:
                        matrix[i][j] = 999;     


def createListOfOnes(size):
    ones = []
    for i in range(size):
        ones.append(1)
    return ones

def startingListFromX(size):
    start = ['x', 0]
    for i in range(size - 2):
        start.append(999)
    return start
    
def shortestPath(matrix,size,checked):
    currentShortest = []
    currentShortest.append(createListOfOnes(size))
    currentShortest.append(startingListFromX(size))
   # print(currentShortest)
    while (checked!=size):
        for i in range(1,size):
            for j in range(size-1,0,-1):
                if (matrix[i][j]==0):
                   pass 
                elif ((matrix[i][j]+currentShortest[1][i])<=currentShortest[1][j]):
                    currentShortest[1][j]=(matrix[i][j]+currentShortest[1][i])
                    currentShortest[0][j]=i
        checked+=1
    #noZero = [x, a, b, c, d, e, f, g, y]
    pathFromY = ['y']
    prev = currentShortest[0][w - 1] #currently 0
    while (prev!=1):
        pathFromY.append(noZero[prev-1])
        prev=currentShortest[0][prev]
    path2Y=['x']
    for i in range(len(pathFromY)-1,-1,-1):
        path2Y.append(pathFromY[i])
    currentShortest[0]=path2Y
    currentShortest[1]=currentShortest[1][w - 1]
    return currentShortest;

finalAnswer = shortestPath(matrix,len(matrix),0)
print("The shortest path is", finalAnswer[0])
print("The distance is", finalAnswer[1])

file.close()
```
